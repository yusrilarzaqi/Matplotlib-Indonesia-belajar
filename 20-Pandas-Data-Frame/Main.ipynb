```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

print(matplotlib.__version__)
print(pd.__version__)
```

```python
df = pd.DataFrame({
  'nama': ['Yusril', 'Bimo', 'Dimas', 'Adam', 'Anis'],
  'usia': [19, 18, 19, 19, 18],
  'gender': ['L', 'L', 'L', 'L', 'P'],
  'propinsi': ['jateng', 'jatim', 'jabar', 'jateng', 'jateng'],
  'n_anak': [0, 0, 0, 0, 0],
  'n_pets': [2, 0, 4, 3, 1]
})
df
```

# Scatter

```python
df.plot(kind='scatter', x='n_anak', y='n_pets', color='red')

plt.title('Scatter Plot')
plt.xlabel('Jumlah Anak')
plt.ylabel('Jumlah Pets')
plt.show()
```

# Histogram

```python
df[['usia']].plot(kind='hist', bins=[x for x in range(1, 30)])

plt.title('Histogram')
plt.xlabel('Usia')
plt.ylabel('Jumlah')
plt.show()
```

# Line Plot

```python
df.plot(kind='line', x='nama', y='n_pets', color='green')

plt.title('Line Plot')
plt.ylabel('Jumlah pets')
plt.show()
```

# Multiple Line Plot

```python
df.plot(kind='line', x='nama',y='n_anak',color='green')
df.plot(kind='line', x='nama',y='n_pets',color='red')

plt.title('Line Plot')
plt.ylabel('Jumlah')
plt.show()
```

```python
ax = plt.gca()

df.plot(kind='line', x='nama',y='n_anak',color='green', ax=ax)
df.plot(kind='line', x='nama',y='n_pets',color='red', ax=ax)

plt.title('Line Plot')
plt.ylabel('Jumlah')
plt.show()
```

# Bar Plot

```python
df.plot(kind='bar', x='nama', y='usia')

plt.title('Bar Plot')
plt.ylabel('Usia')
plt.show()
```

```python
df.groupby('propinsi')['nama'].count().plot(kind='bar')

plt.title('Bar Plot')
plt.ylabel('Jumlah')
plt.show()
```

# Stacked bar plot dengan dua label group by

```python
df.groupby(['propinsi', 'gender']).size().unstack().plot(kind='bar', stacked=True)

plt.title('Bar Plot')
plt.ylabel('Jumlah')
plt.show()
```

```python
df.groupby(['gender', 'propinsi']).size().unstack().plot(kind='bar', stacked=True)

plt.title('Bar Plot')
plt.ylabel('Jumlah')
plt.show()
```
