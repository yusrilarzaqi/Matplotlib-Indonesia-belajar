# Visualisai Data yang tersimpan pada Pandas Data Frame (Bagian3)

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

print(matplotlib.__version__)
print(pd.__version__)
```

## **Sample Dataset**

```python
df = pd.read_csv('./2020.csv')
df.head()
```

```python
df.rename(columns={
  'Country name' : 'Country',
  'Perceptions of corruption': 'Corruption',
  'Freedom to make life choices': 'Freedom'
  },inplace=True)
df.columns
```

# **Bar Plot**

```python
df[0:10].plot(
  x='Country',
  y=['Corruption', 'Freedom', 'Generosity', 'Social support'],
  kind='bar'
)

plt.title('Contoh Bar plot')
plt.ylabel('Nilai')
plt.show()
```

# **Line Plot**

```python
df[:5].plot(
  x='Country',
  y=['Corruption', 'Freedom', 'Generosity', 'Social support'],
  kind='line'
)

plt.title('Contoh Line Plot')
plt.ylabel('Nilai')
plt.show()
```

# **Box Plot**

```python
df[:20].plot(
    y=['Corruption', 'Freedom', 'Generosity', 'Social support'],
    kind='box'
)

plt.title('Contoh Box plot')
plt.xlabel('Parameter')
plt.ylabel('Nilai')
plt.show()
```

# **Scatter Plot**

```python
df[:].plot(
  kind='scatter',
  x='Corruption',
  y='Freedom'
)

plt.xlim((0, 1))
plt.ylim((0, 1))

plt.title('Contoh Sccater Plot')
plt.show()
```

# **Histogram**

```python
df.plot(
  x='Country',
  y='Corruption',
  kind='hist',
  bins=40
)
plt.title('Contoh Histogram')
plt.xlabel('Nilai')
plt.show()
```

```python
df.plot(
  x='Country',
  y=['Corruption', 'Freedom', 'Generosity', 'Social support'],
  subplots=True,
  kind='hist',
  layout=(2, 2)
)

plt.suptitle('Contoh Subplot')
plt.tight_layout()
plt.show()
```
